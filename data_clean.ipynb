```python
min_minutes = 30

def preprocess(df):
    minutes_ = df.iloc[:, 3] - df.iloc[:, 2]
    df['tripduration'] = minutes_.dt.total_seconds()/60
    df.iloc[:, -1] = round(df.iloc[:, -1])
    df_short = df[df.iloc[:, -1] >= min_minutes].copy()
    
    return df_short
```

```python
preprocess(df_202206)
```

```python
files_to_merge = [file for file in glob.glob(r'data/*.csv')]
```

```python
files_to_merge
```

```python
number_of_cols = list()

for file in files_to_merge:
    df = pd.read_csv(file, nrows=0)
    number_of_cols.append(df.columns.size)
    
print(f"Maximum number of columns: {max(number_of_cols)}")
print(f"Maximum number of columns: {max(number_of_cols)}")
```

```python
col_names = pd.concat([pd.read_csv(file, nrows = 0) for file in files_to_merge], 
                      ignore_index = True).columns
col_names
```

```python
data_list = list()

for file in files_to_merge:

    df = pd.read_csv(file, parse_dates=['started_at', 'ended_at'])

    # Preprocess each dataframe
    processed_df = preprocess(df)
    
    # Append each dataframe to the data_list
    data_list.append(processed_df)
```

```python
# Conso list of dataframes into one dataframe
conso_df = pd.concat(data_list)

# Drop duplicates
conso_df.drop_duplicates(inplace=True)
```
